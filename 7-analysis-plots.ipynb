```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
sns.set_context("paper")
```

```python
data = pd.read_csv('output/results.csv')
data.loc[:, 'f1'] = data.loc[:, 'f1'].apply(lambda x: float(x.strip('%')) / 100)
```

```python
data_part = data[data.radius==1000][['incident_interval', 'aperture_size', 'time_step', 'f1']]

# Plot the responses for different events and regions
sns.lineplot(x="incident_interval", y="f1", hue="time_step", style="aperture_size", data=data_part)

# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

print('Variation of F1 with different Incident Intervals.')
```
